```python
import pandas as pd

df = pd.read_excel("FPL_allplayers_last3_gwk_6.xlsx")

print("Rows and columns:", df.shape)
print("\nFirst 5 rows:")
print(df.head())

print("\nColumn info")
df.info()

print("\nColumn Names")
print(df.columns.tolist())

#check and fill missing values
print(df.isnull().sum())
df.fillna(0, inplace=True)

print("\nNumeric Columns Summary stats")
print(df.describe())

print("\nTop 10 Players by Points:")
print(df[['Player', 'Team', 'Pos', 'Points']].sort_values('Points', ascending=False).head(10))

print("\nAvg Points per Position:")
print(df.groupby('Pos')['Points'].mean().sort_values(ascending=False))

print("\nAvg Goals per Position:")
print(df.groupby('Pos')['Goals'].mean().sort_values(ascending=False))

print("\nTotal Points by Team:")
print(df.groupby('Team')['Points'].sum().sort_values(ascending=False))

print("\nAnalysis complete!")
```
