```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("Boston.csv")

print(df.head())

X = df['rm']
y = df['medv']

X = np.array(X)
y = np.array(y)

x_mean = np.mean(X)
y_mean = np.mean(y)

b1 = np.sum((X - x_mean) * (y - y_mean)) / np.sum((X - x_mean)**2)

b0 = y_mean - b1 * x_mean

print(f"\nEquation of line: y = {b0:.2f} + {b1:.2f}x")

y_pred = b0 + b1 * X

mse = np.mean((y - y_pred)**2)
ss_total = np.sum((y - np.mean(y))**2)
ss_residual = np.sum((y - y_pred)**2)
r2 = 1 - (ss_residual / ss_total)

print("\nModel Performance:")
print(f"Mean Squared Error: {mse:.2f}")
print(f"RÂ² Score: {r2:.3f}")

example_rm = 6  # e.g., house with 6 rooms
predicted_price = b0 + b1 * example_rm
print(f"\nPredicted house price for {example_rm} rooms: {predicted_price:.2f}")


plt.figure(figsize=(8,6))

plt.scatter(X, y, color='blue', label='Actual Data')

plt.plot(X, y_pred, color='red', label='Regression Line')

plt.title('Simple Linear Regression (RM vs MEDV)')
plt.xlabel('Average number of rooms per dwelling (RM)')
plt.ylabel('Median value of homes (MEDV, $1000s)')
plt.legend()
plt.grid(True)
plt.show()
```
